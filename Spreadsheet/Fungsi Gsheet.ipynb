# Find dan Search

Gunakan FIND jika Anda peduli dengan huruf besar/kecil.

# Edate
Fungsi EDATE digunakan untuk menghitung tanggal jatuh tempo atau tanggal di masa depan/masa lalu dengan menambahkan atau mengurangi jumlah bulan dari tanggal yang diberikan.



```python
TEXT(EDATE(TODAY(), -1), "MMM")

"""
Fungsi diatas menampilkan bulan dalam bentukan "MMM" karena -1 maka yang akan dikeluarkan adalah bulan lalu
bulan lalu)
"""

TEXT(EDATE(TODAY(), -2), "MMM") # akan keluar b2 bulan lalau


TEXT(EDATE(TODAY(), -2), "MMM") # akan keluar last 3 month
```
